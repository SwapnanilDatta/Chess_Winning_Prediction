```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error,classification_report
```

```python
data=pd.read_csv('cleaned_data.csv')
data['opening_name']
```

```python




# Define a mapping dictionary to normalize category names
mapping = {
    r'Philidor Defense.*': 'Philidor Defense',
    r'Queen\'s Pawn Game.*': 'Queen\'s Pawn Game',
    r'Queen\'s Pawn.*': 'Queen\'s Pawn',
    r'Sicilian Defense.*': 'Sicilian Defense',
    r'French Defense.*': 'French Defense',
    r'Italian Game.*': 'Italian Game',
    r'King\'s Pawn Game.*': 'King\'s Pawn Game'
}

# Apply the mapping to normalize the category names
for pattern, replacement in mapping.items():
    data['opening_name'] = data['opening_name'].str.replace(pattern, replacement, regex=True)

print(data['opening_name'])
```

```python
data.to_csv('cleaned.csv')
```

```python
data = pd.read_csv('cleaned.csv')
le = LabelEncoder()
data['opening_name'] = le.fit_transform(data['opening_name'])
data['winner'] = data['winner'].map({'white': 1, 'black': 0, 'draw': 2})
```

```python
X = data[['white_rating', 'black_rating', 'opening_name']]
y_turns = data['turns']  # Regression target
y_winner = data['winner']  # Classification target
```

### Turns Calculating using Regression


```python
X_train, X_test, y_turns_train, y_turns_test = train_test_split(X, y_turns, test_size=0.2, random_state=42)
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
   
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_turns_train) # Train model

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Evaluate Train and Test dataset
    model_train_mae , model_train_rmse, model_train_r2 = evaluate_model(y_turns_train, y_train_pred)

    model_test_mae , model_test_rmse, model_test_r2 = evaluate_model(y_turns_test, y_test_pred)

    
    print(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
    print("- R2 Score: {:.4f}".format(model_train_r2))

    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
    print("- R2 Score: {:.4f}".format(model_test_r2))
    
    print('='*35)
    print('\n')
```

```python
from sklearn.pipeline import Pipeline
reg_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor())
])

reg_param_grid = {
    'model__n_estimators': [50, 100, 150],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}

reg_grid_search = GridSearchCV(reg_pipeline, reg_param_grid, cv=5, scoring='neg_mean_absolute_error', verbose=1)
reg_grid_search.fit(X_train, y_turns_train)

# Best Regression Model
best_reg_model = reg_grid_search.best_estimator_
print("Best Regression Parameters:", reg_grid_search.best_params_)

# Regression Evaluation
y_turns_pred = best_reg_model.predict(X_test)
print("Mean Absolute Error for Turns:", mean_absolute_error(y_turns_test, y_turns_pred))
```

```python
import pickle
from sklearn.ensemble import RandomForestRegressor



# Pickle the trained model to a file
with open('turns_model.pkl', 'wb') as file:
    pickle.dump(best_reg_model, file)
```

### Classifier Using for Winner Prediction

```python
_, _, y_winner_train, y_winner_test = train_test_split(X, y_winner, test_size=0.2, random_state=42)
```

```python
clf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier())
])

clf_param_grid = {
    'model__n_estimators': [50, 100, 150],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}

clf_grid_search = GridSearchCV(clf_pipeline, clf_param_grid, cv=5, scoring='accuracy', verbose=1)
clf_grid_search.fit(X_train, y_winner_train)

# Best Classification Model
best_clf_model = clf_grid_search.best_estimator_
print("Best Classification Parameters:", clf_grid_search.best_params_)

# Classification Evaluation
y_winner_pred = best_clf_model.predict(X_test)
print(classification_report(y_winner_test, y_winner_pred))
```

```python
with open('winner_model.pkl', 'wb') as file:
    pickle.dump(best_clf_model, file)
```
